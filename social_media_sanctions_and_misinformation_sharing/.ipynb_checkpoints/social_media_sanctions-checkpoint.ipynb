```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler 
import pingouin as pg
```

```python
filepath = './mosleh_et_al_data.csv'
df= pd.read_csv(filepath)
```

## Twitter Suspensions after the 2020 Election

### Crosstab

```python
crosstab = pd.crosstab(df['politics_hashtag'], df['suspended'])
print(crosstab)
```

The cross tabulation rows show for each politics_hashtag the frequency of a profile using it being suspended. Let's check the ratio between them:

```python
ratio = crosstab[1][1] / crosstab[1][0]
print(f"It is {ratio:.2f} times more likely for an account using hastag #Trump2020 to be suspended compared to #VoteBidenHarris2020.")
```

### $\chi^2$ test
Our null hypothesis is that there is no relationship between the hashtag a profile uses and them being suspended.
All of the chi-square test assumptions are met, so we proceed:

```python
crosstab_freq = pd.crosstab(df['politics_hashtag'], df['suspended'])
chi2, pvalue, dof, expected = stats.chi2_contingency(crosstab_freq)
print('Chi squared statistic: ',chi2)
print('P-value: ', pvalue)
print('Degrees of freedom: ', dof)
```

- P-value is much smaller than 0.05 so the null hypothesis is rejected. We can conclude that there is a statistically significant association between the hashtag used and the likelihood of account suspension.
- Chi squared statistic is large, showing a big difference between the frequencies of suspension in the crosstab and the expected frequencies of suspension if there was no relationship between the hashtag used and the suspension.
- Degrees of freedom is one, logical for a 2x2 crosstab.

### T-test
The zero hypothesis is that the difference in group means is zero. We perform an independent T-test, as the data regarding the hashtag usage and the low quality sharing was drawn from different populations.

```python
df = df['lowqual_pr2019_fc_standardized'].fillna(df['lowqual_pr2019_fc_standardized'].mean())
df = df['lowqual_pr2019_crowd_standardized'].fillna(df['lowqual_pr2019_crowd_standardized'].mean())
#df[['politics_hashtag']] = scaler.fit_transform(df[['politics_hashtag']]) standardize??

t_statistic, p_value = stats.ttest_ind(df['lowqual_pr2019_fc_standardized'], df['politics_hashtag'])
print('T-statistic: ', t_statistic)
print('P-value: ', p_value)
```

```python
t_statistic2, p_value2 = stats.ttest_ind(df['lowqual_pr2019_crowd_standardized'], df['politics_hashtag'])
print('T-statistic: ', t_statistic2)
print('P-value: ', p_value2)
```

Both T-tests run, show a big distance between the means of the 2 groups taken each time (once for the low quality based on the fact-checkers and once for the low quality from crowdsourcing). P-value is very small (<0.001 but typically represented as zero) showing strong evidence that the means of the 2 groups differ and this difference did not come by chance. 

To perform the t-test both metrics should follow a normal distribution. The binary politics_hashtag violates this assumption, and normalizing it whould result in promoting the zeros (the rearer class), so the hashtags for Biden. Let's use the U-test, which is a non parametric test, meaning it doesn't require normality for the data.  
The Mann Whitney U test provides evidence that there is a statistically significant difference in the distribution of your continuous variable between the two groups defined by your binary variable.

Let's perform a T-test, to check if there are significant differences between the low quality determined by fact-checkers and crowdsourcing. We assume that there is no difference in the population means of the groups (H0).

```python
t_statistic, p_value = stats.ttest_ind(df['lowqual_pr2019_fc_standardized'], df['lowqual_pr2019_crowd_standardized'])
print('T-statistic: ', t_statistic)
print('P-value: ', p_value)
```

The p-value is very big (near 1) providing strong evidence that there is no statistically significant difference in the distribution of the low quality data between the groups. Now we have proved that the two plots we generated do not look alike by chance.

## Low-quality News Sharing on Twitter

```python
def hedges_g(cohens_d, n1, n2):
  if not all(isinstance(x, (int, float)) and x > 0 for x in [cohens_d, n1, n2]):
      print("Error: Sample sizes and Cohen's d must be positive numbers.")
      return None
  n = n1 + n2
  if n <= 2:
      print("Error: Total sample size must be greater than 2.")
      return None
  g = cohens_d * (1 - (3 / (4 * (n1 + n2)-9)))
  return g
```

```python
group0 = df[df["politics_hashtag"] == 0]["lowqual_pr2019_fc_standardized"]
group1 = df[df["politics_hashtag"] == 1]["lowqual_pr2019_fc_standardized"]

cohen_d = pg.ttest(group0, group1)['cohen-d']
print('Cohens d: ',cohen_d)
print('Hedges g: ',hedges_g(2.523332,len(group0),len(group1)))

t_statistic, p_value = stats.ttest_ind(group0, group1)
print('T-statistic: ', t_statistic)
print('P-value: ', p_value)
```

While p-value informs us that the difference in the means of the distributions are unlikely due to chance, Cohen's d quantifies the effect size for the difference between two group means. In this case the effect is huge (Cohen's d= 2.52 is huge according to the interpretation).
<p>Let's do the same for the other ratings of low quality:</p>

```python
ratings = ['lowqual_pr2019_fc','lowqual_afm', 'lowqual_mbfc', 'lowqual_lasser2019', 'lowqual_pr2019_crowd', 'lowqual_pr2019_crowdrep']
for r in ratings:
    if (f'{r}_standardized') not in df.columns:
        df[f'{r}_standardized'] = scaler.fit_transform(df[[r]])

def calculate_effect_size(lowqual):
  group0 = df[df["politics_hashtag"] == 0][f'{lowqual}_standardized']
  group1 = df[df["politics_hashtag"] == 1][f'{lowqual}_standardized']
  
  t_test = pg.ttest(group0, group1)

  return pd.Series({
      'Rating': lowqual,
      'T-statistic': t_test['T'].iloc[0],
      'P-value': t_test['p-val'].iloc[0],
      'Cohen\'s d': t_test['cohen-d'].iloc[0],
      'Hedge\'s g': hedges_g(t_test['cohen-d'].iloc[0], len(group0), len(group1)),
  })

results_df = pd.DataFrame(ratings, columns=['Rating'])
results_df = results_df.apply(lambda row: calculate_effect_size(row['Rating']), axis=1)
print(results_df)
```

- For each rating (coming from a different source) of the low quality, the pingouin library provides the Cohen's d measure. Cohen's d is unitless. In the end we can take all the findings with differing units and place them on the same standardized scale, which facilitates comparisons between the ratings given by different sources. 
- The results show that in most cases, the effects magnitude are huge. The only exception is the ratings conducted by the republicans among the 971 participants of the other 4, which is closer to very large (they rated the low qualities between the groups a little more similarly), but still proving the clear difference between the groups of users.
- The sample size is in all cases is bigger than 20. That's why Hedges’ g and Cohen’s d results are roughly equivalent.

### Sharing Lower-quality News Sources and Conservative Ideology

```python
ratings = ['lowqual_pr2019_fc', 'lowqual_lasser2019','lowqual_mbfc','lowqual_afm','lowqual_pr2019_crowd', 'lowqual_pr2019_crowdrep']
politics = ['politics_followed','politics_hashtag','politics_sites1','politics_sites2']
for p in politics:
    if (f'{p}_standardized') not in df.columns and p != 'politics_hashtag':
        df[f'{p}_standardized'] = scaler.fit_transform(df[[p]])
        df = df[f'{p}_standardized'].fillna(df[f'{p}_standardized'].mean())
        
print(df['politics_followed_standardized'],df['politics_followed'])
ratings = [item + "_standardized" for item in ratings]
politics = [item + "_standardized" if item != 'politics_hashtag' else item for item in politics]
df_subset = df[ratings + politics]
correlation_matrix = df_subset.corr(method='pearson')
correlation_matrix = correlation_matrix.loc[ratings, politics]

# Set the index and columns explicitly
rating_politics_correlations = correlation_matrix.copy()
rating_politics_correlations.index.name = 'Low quality news sources'
rating_politics_correlations.columns.name = 'Political ideology'

print("Pearson Correlation Matrix:\n", rating_politics_correlations)
```

```python
plt.figure(figsize=(8, 6)) 
sns.heatmap(correlation_matrix, annot=True, cmap="GnBu", linewidths=.5)
plt.yticks(rotation=0)
plt.show()
```
